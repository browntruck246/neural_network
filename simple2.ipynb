## Define the sigmoid function

The sigmoid function is a mathematical function that maps any input value (real number) to a value between 0 and 1. It's widely used in fields like machine learning and statistics, particularly in logistic regression and neural networks. Here's how it works:

The sigmoid function is defined as:

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

Where:

* **𝑥** is the input.

* **𝑒** is the base of the natural logarithm (approximately 2.718).

## Key characteristics:

* **1 Range:** The output of the sigmoid function is always between 0 and 1, regardless of the input value. This makes it great for tasks requiring probabilities or classifications.

* **2 Shape:** It has an S-shaped curve (sigmoid curve), with steep slopes around 𝑥 = 0. For large positive 𝑥, the output approaches 1. For large negative 𝑥, the output approaches 0.

* **3 Smooth Gradient:** The function is smooth and differentiable, meaning it allows gradient-based optimization methods, which are crucial for training machine learning models.

## Applications:
* **Logistic Regression:** Used to predict probabilities and classify binary outcomes.

* **Neural Networks:** Often used as an activation function to introduce non-linearity, enabling models to capture complex patterns.

The sigmoid function essentially compresses the input into a narrow output range, making it ideal for modeling and decision-making.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

<p style="color: blue; font-size: 30px;"><b>Explain Mean Squared Error Loss</b></p> 

Mean Squared Error (MSE) loss is a common metric used in regression tasks to measure the difference between predicted values and actual values. Here's how it works:

## Formula:
The MSE is calculated as:

$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Where:         

* The actual value$$y_i$$    

* The predicted value $$\hat{y}_i$$     

* The number of data points $$n$$          

## Explanation:
**1 Error Calculation:**

* For each data point, the difference between the actual value and the predicted value is calculated.

* This difference is squared to ensure all errors are positive and to penalize larger errors more heavily.

**2 Averaging:**

* The squared errors are summed up and divided by the number of data points to compute the mean.

## Characteristics:
* Range: MSE is always non-negative, as squared values cannot be negative.

* Interpretation: A smaller MSE indicates better predictions, as the errors are closer to zero.

```python
# Mean squared error loss
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

## Applications:
MSE is widely used in machine learning models, especially for regression tasks, to evaluate the performance of a model. It helps in optimizing the model parameters by minimizing the loss during training.

```python
import numpy as np

# Define the actual and predicted arrays
actual = np.array([3, -0.5, 2, 7])  # Replace with your actual values
predicted = np.array([2.5, 0.0, 2, 8])  # Replace with your predicted values

# Calculate the Mean Squared Error
mse = np.mean((actual - predicted)**2)

print(f"Mean Squared Error: {mse}")
```

## Define the neural network structure

```python
input_neurons = 3  # Number of features in the input
hidden_neurons = 2 # Number of neurons in the hidden layer
output_neurons = 1 # Binary output (0 or 1)
```

# Initialize weights and biases randomly

```python
weights_input_hidden = np.random.rand(input_neurons, hidden_neurons)
weights_hidden_output = np.random.rand(hidden_neurons, output_neurons)
bias_hidden = np.random.rand(1, hidden_neurons)
bias_output = np.random.rand(1, output_neurons)

print( "Weights Input Hidden : \n"  + str(weights_input_hidden) + "\n")
print( "Weights Hidden Output : \n"  + str(weights_hidden_output) + "\n")
print( "Bias Hidden : \n" + str(bias_hidden) + "\n")
print( "Bias Output : \n" + str(bias_output) + "\n")
```

# Training parameters

```python
learning_rate = 0.1
epochs = 1000
```

# Print A Loop Function

```python
def check_and_print(loopNum, loopPrint, message):
    # Check if loopNum is in loopPrint
    if loopNum in loopPrint:
        # Print the message
        print(message)
```

```python
loopNum = 5
loopPrint = [1, 2, 3, 5]
message = "Match found!"
check_and_print(loopNum, loopPrint, message)
```

# Training the network

```python
for epoch in range(epochs):

    loopPrint = [0]
    check_and_print( epoch, loopPrint, "---------------------------------------- LOOP NUMBER : " + str(epoch)   + "----------------------------------------" )

    # Forward pass
    check_and_print( epoch, loopPrint, "______________________________ Forward Pass ______________________________ " )
    
    check_and_print( epoch, loopPrint, " X : \n" + str(X)  + "\n")
    check_and_print( epoch, loopPrint, " weights_input_hidden: \n" + str(weights_input_hidden)  + "\n")
    check_and_print( epoch, loopPrint, " np.dot(X, weights_input_hidden) : \n" + str(np.dot(X, weights_input_hidden))  + "\n")
    check_and_print( epoch, loopPrint, " bias_hidden : \n" + str(bias_hidden)  + "\n")
    hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
    check_and_print( epoch, loopPrint, " hidden_layer_input : \n" + str(hidden_layer_input) + "\n" )

    check_and_print( epoch, loopPrint, "______________________________ Sigmoid ______________________________" )

    check_and_print( epoch, loopPrint, " Hidden Layer Input : \n" + str(hidden_layer_input)  + "\n")
    hidden_layer_output = sigmoid(hidden_layer_input)
    check_and_print( epoch, loopPrint, " sigmoid(hidden_layer_input) : \n" + str(hidden_layer_output)  + "\n")

    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " hidden_layer_output : \n" + str(hidden_layer_output)  + "\n")
    check_and_print( epoch, loopPrint, " weights_hidden_output: \n" + str(weights_hidden_output)  + "\n")
    check_and_print( epoch, loopPrint, " np.dot(hidden_layer_output, weights_hidden_output) : \n" + str(np.dot(hidden_layer_output, weights_hidden_output))  + "\n")
    check_and_print( epoch, loopPrint, " bias_hidden : \n" + str(bias_hidden)  + "\n")
    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
    check_and_print( epoch, loopPrint, " output_layer_input : \n" + str(output_layer_input)  + "\n")

    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " output_layer_input : \n" + str(output_layer_input)  + "\n")
    output = sigmoid(output_layer_input)
    check_and_print( epoch, loopPrint, " output : \n" + str(output)  + "\n")

    check_and_print( epoch, loopPrint, "_______________________________ Compute Loss __________________________" )
    
    # Compute loss
    check_and_print( epoch, loopPrint, " y : \n" + str(y)  + "\n")
    check_and_print( epoch, loopPrint, " output : \n" + str(output)  + "\n")
    loss = mse_loss(y, output)
    check_and_print( epoch, loopPrint, " loss : \n" + str(loss)  + "\n")

    check_and_print( epoch, loopPrint, "_______________________________ Backward Pass __________________________" )
    
    # Backward pass
    check_and_print( epoch, loopPrint, " y : \n" + str(y)  + "\n")
    check_and_print( epoch, loopPrint, " output : \n" + str(output)  + "\n")
    error_output = y - output
    check_and_print( epoch, loopPrint, " error_output : \n" + str(error_output)  + "\n")

    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " error_output : \n" + str(error_output)  + "\n")
    check_and_print( epoch, loopPrint, " output : \n" + str(output)  + "\n")
    check_and_print( epoch, loopPrint, " sigmoid_derivative(output) : \n" + str(sigmoid_derivative(output))  + "\n")
    delta_output = error_output * sigmoid_derivative(output)
    check_and_print( epoch, loopPrint, " delta_output : \n" + str(delta_output)  + "\n")

    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " weights_hidden_output.T : \n" + str(weights_hidden_output.T)  + "\n")
    check_and_print( epoch, loopPrint, " delta_output.dot(weights_hidden_output.T) : \n" + str(delta_output.dot(weights_hidden_output.T))  + "\n")
    error_hidden = delta_output.dot(weights_hidden_output.T)
    check_and_print( epoch, loopPrint, " error_hidden : \n" + str(error_hidden)  + "\n")

    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " error_hidden : \n" + str(error_hidden)  + "\n")
    check_and_print( epoch, loopPrint, " hidden_layer_output : \n" + str(hidden_layer_output)  + "\n")
    check_and_print( epoch, loopPrint, " sigmoid_derivative(hidden_layer_output) : \n" + str(sigmoid_derivative(hidden_layer_output))  + "\n")
    delta_hidden = error_hidden * sigmoid_derivative(hidden_layer_output)
    check_and_print( epoch, loopPrint, " delta_hidden : \n" + str(delta_hidden)  + "\n")

    check_and_print( epoch, loopPrint, "____________________________________ Update weights and biases ________________________________" )
    # Update weights and biases
    check_and_print( epoch, loopPrint, " hidden_layer_output.T : \n" + str(hidden_layer_output.T)  + "\n")
    check_and_print( epoch, loopPrint, " delta_output : \n" + str(delta_output)  + "\n")
    check_and_print( epoch, loopPrint, " hidden_layer_output.T.dot(delta_output) : \n" + str( hidden_layer_output.T.dot(delta_output) )  + "\n")
    check_and_print( epoch, loopPrint, " learning_rate : \n" + str(hidden_layer_output.T.dot(learning_rate))  + "\n")
    weights_hidden_output += hidden_layer_output.T.dot(delta_output) * learning_rate
    check_and_print( epoch, loopPrint, " weights_hidden_output : \n" + str( weights_hidden_output )  + "\n")
    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " X.T : \n" + str(X.T)  + "\n")
    check_and_print( epoch, loopPrint, " delta_hidden : \n" + str(delta_hidden)  + "\n")
    check_and_print( epoch, loopPrint, " X.T.dot(delta_hidden) : \n" + str(X.T.dot(delta_hidden))  + "\n")
    check_and_print( epoch, loopPrint, " learning_rate: \n" + str(learning_rate)  + "\n")
    weights_input_hidden += X.T.dot(delta_hidden) * learning_rate
    check_and_print( epoch, loopPrint, " weights_input_hidden : \n" + str(weights_input_hidden)  + "\n")
    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " learning_rate : \n" + str(learning_rate)  + "\n \n")

    check_and_print( epoch, loopPrint, " delta_output : \n" + str(delta_output)  + "\n")
    check_and_print( epoch, loopPrint, " np.sum(delta_output, axis=0, keepdims=True) : \n" + str(np.sum(delta_output, axis=0, keepdims=True))  + "\n")
    bias_output += np.sum(delta_output, axis=0, keepdims=True) * learning_rate
    check_and_print( epoch, loopPrint, " bias_output : \n" + str(bias_output)  + "\n")

    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    check_and_print( epoch, loopPrint, " learning_rate : \n" + str(learning_rate)  + "\n \n")

    check_and_print( epoch, loopPrint, " delta_hidden : \n" + str(delta_hidden)  + "\n")
    check_and_print( epoch, loopPrint, " np.sum(delta_hidden, axis=0, keepdims=True) : \n" + str(np.sum(delta_hidden, axis=0, keepdims=True))  + "\n")
    bias_hidden += np.sum(delta_hidden, axis=0, keepdims=True) * learning_rate
    check_and_print( epoch, loopPrint, " bias_hidden : \n" + str(bias_hidden)  + "\n")

    check_and_print( epoch, loopPrint, "_____________________________________________________________________" )

    # Print loss every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss:.4f}")



# Final output after training
print("\nTraining complete.")
```

<p style="color: blue; font-size: 30px;"><b>Compute Dot Product</b></p>

## How to Compute It
There are two main ways to compute the dot product:

### **1. Geometric Definition:**
The dot product measures the projection of one vector onto another. Mathematically, it's expressed as: $$a ⋅ b = |a| × |b| × \cos(θ)$$

Where:

* ∣𝑎∣ and ∣𝑏∣ are the magnitudes (lengths) of the vectors 𝑎 and 𝑏
* 𝜃 is the angle between the vectors.
* cos(𝜃) is the cosine of that angle.

**Example**         
Let's illustrate the formula for the dot product of two vectors, a ⋅ b = |a| × |b| × cos(θ), with numbers.

Suppose vector a has a magnitude |a| = 5, and vector b has a magnitude |b| = 3. Let the angle θ between the two vectors be 60°. Using the formula:

a ⋅ b = |a| × |b| × cos(θ)

|a| = 5, |b| = 3, and cos(60°) = 0.5

Substitute these values: a ⋅ b = 5 × 3 × 0.5

Perform the calculation: a ⋅ b = 15 × 0.5 = 7.5

So, the dot product of a and b in this example is 7.5.

**Converting two vectors into two arrays**

To convert this information into arrays, we can compute the components of the vectors a and b based on their magnitudes and the angle θ.

**Breakdown:**
* 1. Vector a has a magnitude |a| = 5. Assume it lies along the x-axis for simplicity: **a = [5, 0]** (x-component is 5, y-component is 0).
 

* 2. Vector b has a magnitude |b| = 3 and forms an angle of θ = 60° with vector a. Use trigonometric functions to calculate its components:    

x-component: bₓ = |b| × cos(θ)    

y-component: bᵧ = |b| × sin(θ)      

Substituting values:   

bₓ = 3 × cos(60°) = 3 × 0.5 = 1.5    

bᵧ = 3 × sin(60°) = 3 × √3/2 ≈ 3 × 0.866 = 2.598    

Therefore, b = [1.5, 2.598].     

Final Output:   
a = [5, 0]  

b ≈ [1.5, 2.598]   

Now the vectors are expressed as arrays!    

### **2. Algebraic Definition:**

If the vectors are written in component form as: $$ a = [a_1, a_2, ..., a_n], \ b = [b_1, b_2, ..., b_n] $$

The dot product is calculated by multiplying corresponding components and summing them: $$ a ⋅ b = a_1b_1 + a_2b_2 + ... + a_nb_n $$

**Steps:***     
aₓ = 5, aᵧ = 0 (components of vector a)   

bₓ = 1.5, bᵧ ≈ 2.598 (components of vector b)    

Substitute into the formula: a ⋅ b = (5 × 1.5) + (0 × 2.598)    

Calculate: a ⋅ b = 7.5 + 0 = 7.5

* A matrix multiplication between the **input features (X)** and **the weight matrix (weights_input_hidden)**.

* Adding the **bias (bias_hidden)** to each **result for the neurons in the hidden layer**.

* X is the input data. It's a 1x3 matrix (a single sample with 3 features): $$ X = \begin{bmatrix} 0.5488135 & 0.71518937 & 0.60276338 \end{bmatrix} $$

$$ np.dot(X, WeightsInputHidden) = \begin{bmatrix} 0.5488135 & 0.71518937 & 0.60276338 \end{bmatrix} \cdot \begin{bmatrix} 0.84725174 & 0.6235637 \\ 0.38438171 & 0.29753461 \\ 0.05671298 & 0.27265629 \end{bmatrix} $$
